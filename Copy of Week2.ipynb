## **List**

*   List is a non-homogeneous data structure which stores the element in single row and multiple rows and columns
*   List can be represented by [ ]

*   List can be created using **list()** function
*   Example: [1,2,3,4,5]








## **Tuple**

*   Tuple is also a non-homogenous data structure which stores single rows and multiple rows and columns.
*   Tuple can be represented by ( )

*   Tuple can be created using **tuple()** function
*   Example: (1,2,3,4,5)





## **Set**

*   Set data structure is also non-homogeneous data structure but stores in single row.

*   Set can be represented by { }
*   Set can be created using **set()** function


*   Example: {1,4,5,3,2}



## **Dictionary**

*   Dictionary is also a non-homogenous data structure which store key value pairs
*   Dictionary can be represented by { }

*   Dictionary can be created using **dict()** function
*   Dictionary is ordered were as Set is unordered

*   Example: {1,2,3,4,5}

**11. Write about evaluation reports and types of it**

The evaluation design matrix is an essential tool for planning and organizing an evaluation. It is simply a table with one row for each evaluation question and columns that address evaluation design issues, such
as data collection methods, data sources, analysis methods, criteria for comparisons, etc. The design
matrix links each evaluation question to the means for answering that question

There are different types of evaluation matrices available to test a model which are sa follows:


*   Classification accuracy
*   logaritmic loss
